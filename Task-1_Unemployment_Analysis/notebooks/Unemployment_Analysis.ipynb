## Exploratory Data Analysis (EDA)

### Key Findings

- Calculated the average unemployment rate.

- Analyzed yearly unemployment trends.

```python
# ==========================================
# Graph Settings
# ==========================================

plt.style.use("ggplot")

sns.set_style("whitegrid")

plt.rcParams["figure.figsize"] = (10,6)
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df1["Estimated Unemployment Rate (%)"],
    bins=20,
    kde=True
)

plt.title("Distribution of Unemployment Rate")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Frequency")

plt.show()
```

### Observation

- Most unemployment rates are concentrated in the lower range.
- A few observations have very high unemployment rates.
- The distribution is slightly right-skewed.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x=df1["Estimated Unemployment Rate (%)"]
)

plt.title("Box Plot of Unemployment Rate")

plt.show()
```

### Observation

- Several outliers are visible.
- Some states experienced exceptionally high unemployment.

```python
monthly_data = df1.groupby("Date")["Estimated Unemployment Rate (%)"].mean().reset_index()

plt.figure(figsize=(15,6))

plt.plot(
    monthly_data["Date"],
    monthly_data["Estimated Unemployment Rate (%)"],
    marker="o"
)

plt.title("Monthly Average Unemployment Rate")
plt.xlabel("Date")
plt.ylabel("Average Unemployment Rate (%)")

plt.xticks(rotation=45)

plt.show()
```

### Observation

- Unemployment changes over time.
- A sharp rise is visible during the COVID-19 period.

```python
top10 = df1.groupby("Region")["Estimated Unemployment Rate (%)"] \
           .mean() \
           .sort_values(ascending=False) \
           .head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top10.values,
    y=top10.index
)

plt.title("Top 10 States by Average Unemployment Rate")
plt.xlabel("Average Unemployment Rate (%)")
plt.ylabel("State")

plt.show()
```

### Observation

- These states recorded the highest average unemployment.
- The differences between states are clearly visible.

```python
bottom10 = df1.groupby("Region")["Estimated Unemployment Rate (%)"] \
              .mean() \
              .sort_values() \
              .head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=bottom10.values,
    y=bottom10.index
)

plt.title("Top 10 States with Lowest Unemployment")

plt.xlabel("Average Unemployment Rate (%)")
plt.ylabel("State")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.barplot(
    data=df1,
    x="Area",
    y="Estimated Unemployment Rate (%)"
)

plt.title("Rural vs Urban Unemployment")

plt.show()
```

### Observation

- Compare the average unemployment in rural and urban areas.
- Discuss which area has the higher unemployment rate.

```python
area = df1["Area"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    area,
    labels=area.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Distribution of Rural and Urban Data")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df1.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

- Strong positive or negative correlations are visible.
- Identify relationships between unemployment, employment, and labour participation.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df1,
    x="Estimated Employed",
    y="Estimated Unemployment Rate (%)",
    hue="Area"
)

plt.title("Employment vs Unemployment")

plt.show()
```

```python
sns.pairplot(
    df1[
        [
            "Estimated Unemployment Rate (%)",
            "Estimated Employed",
            "Estimated Labour Participation Rate (%)"
        ]
    ]
)

plt.show()
```

```python
month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

plt.figure(figsize=(12,6))

sns.barplot(
    data=df1,
    x="Month",
    y="Estimated Unemployment Rate (%)",
    order=month_order
)

plt.xticks(rotation=45)

plt.title("Average Unemployment by Month")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df1,
    x="Year",
    y="Estimated Unemployment Rate (%)"
)

plt.title("Average Unemployment by Year")

plt.show()
```

```python
covid = df2.groupby("Date")["Estimated Unemployment Rate (%)"].mean().reset_index()

plt.figure(figsize=(15,6))

plt.plot(
    covid["Date"],
    covid["Estimated Unemployment Rate (%)"],
    marker="o",
    color="red"
)

plt.title("COVID-19 Unemployment Trend")

plt.xlabel("Date")
plt.ylabel("Average Unemployment Rate (%)")

plt.xticks(rotation=45)

plt.show()
```

### Observation

- Observe how unemployment changed during the COVID-19 months.
- Identify the peak unemployment period.

```python
fig = px.scatter_geo(
    df2,
    lat="latitude",
    lon="longitude",
    color="Estimated Unemployment Rate (%)",
    size="Estimated Unemployment Rate (%)",
    hover_name="Region",
    title="Unemployment Rate Across India"
)

fig.show()
```

```python
covid = df2.groupby("Date")["Estimated Unemployment Rate (%)"].mean().reset_index()

plt.figure(figsize=(15,6))

plt.plot(
    covid["Date"],
    covid["Estimated Unemployment Rate (%)"],
    marker="o",
    color="red"
)

plt.title("COVID-19 Unemployment Trend")
plt.xlabel("Date")
plt.ylabel("Average Unemployment Rate (%)")

plt.xticks(rotation=45)

plt.savefig("../images/covid_trend.png", dpi=300, bbox_inches="tight")

plt.show()
```
